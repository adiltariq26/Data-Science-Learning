# Creating a Conditional Column from More Than 2 Choices np.select ()

```python
import numpy as np 
```

```python
# create an array based o multiple price tiers price (+2 choices )
# create "conditions" and "values" variables
conditions = [
    (df_laptops['Price_euros'] > 3000),
    (df_laptops['Price_euros'] > 2000) & (df_laptops['Price_euros'] <= 3000),
    (df_laptops['Price_euros'] > 800) & (df_laptops['Price_euros'] <= 2000),
    (df_laptops['Price_euros'] <= 800)
]
choices = ['Too Expensive','Expensive','Affordable','Cheap']
```

```python
# set it to new column 
df_laptops['Price_tier'] = np.select(conditions, choices,default='Unknown')
```

```python
# display dataframe
df_laptops
# df_laptops['Price_tier'].head(6)
```

```python
# value counts of our condition
df_laptops.value_counts('Price_tier')
```

```python
import pandas as pd
df_laptops = pd.read_csv('laptop_price.csv')
df_laptops.head(4)
```

```python
# create an array based array based on screen size tiers (+2 choices)
# Too big > 16 , big>14 , small >12 , too small <12 
# create conditions and values variables
conditions = [
    (df_laptops['Inches'] >16),
    (df_laptops['Inches'] >14 ) &  (df_laptops['Inches'] <= 16 ),
    (df_laptops['Inches'] >12 ) &  (df_laptops['Inches'] <= 14 ),
    (df_laptops['Inches'] <12 )
]
values = ['Too Big','Big','Small','Too Small']
```

```python
# set it to new column
np.select(conditions,values,default='Unknown')
df_laptops['Screen Size'] = np.select(conditions,values,default='Unknown')
```

```python
# display dataframe
df_laptops.head(5)
```

```python
# value counts
df_laptops.value_counts('Screen Size')
```
