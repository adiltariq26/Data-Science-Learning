# Drop duplicates
- Used to remove duplicates n 2 or more Columns

```python
import pandas as pd 
```

```python
df_laptops = pd.read_csv('laptop_price.csv')
```

```python
df_laptops.head(3)
```

```python
# drop duplicate 
df_laptops.drop_duplicates(['Company'])[['Company','Price_euros']].value_counts('Company')
```

```python
df_laptops.sort_values(['Company','Inches'])
```

```python
# cheapest : keep = first
# df_laptops.drop_duplicates(['Company'],keep = first )
```

```python
df_laptops.drop_duplicates(['Company'] , keep = 'last')[['Company','Inches']]
```

## Excercise 1
- Task: Find the biggest and smallest screen size in laptops of each company using the sort_values() and duplicated() methods (keep "first", "last" and False)

```python
import pandas as pd 
```

```python
df_laptops = pd.read_csv('laptop_price.csv')
```

```python
df_laptops.sort_values(['Company','Inches']).value_counts(['Company','Inches'])
```

```python
df_laptops.drop_duplicates(['Company','Inches'],keep = 'first')[['Company','Inches']]
```

```python
df_laptops.drop_duplicates(['Company','Inches'],keep = 'last')[['Company','Inches']]
```
