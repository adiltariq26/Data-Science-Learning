```python
import pandas as pd
```

```python
df = pd.read_csv('players_20.csv')
```

```python
# set index
df.set_index('short_name', inplace = True )
```

```python
# selecting Columns
df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

## Selecting with Single Value
- loc[row_label,column_label]

```python
# get all data about messi
df.loc['L. Messi']
```

```python
# get height of Messi 
df.loc['L. Messi','height_cm']
```

```python
# Get weight of Cristiano Ronaldo 
df.loc['Cristiano Ronaldo','weight_kg']
```

```python
# get all rows inside 'height_cm' column
# kisi column kei all rows ki values maloom krne kei liey index label kei start pr colon use kr lu
df.loc[:,'height_cm']
```

```python
# kisi Row Index kei all columns ki values maloom krne kei liey index label kei end pr colon use kr lu
df.loc['L. Messi',:]
```

## Selecting with List of Values

```python
# get the height of messi and ronaldo
players = ['L. Messi','Cristiano Ronaldo']
df.loc[players,'height_cm']
```

```python
# get the weight and height of Messi 
attribute = ['height_cm','weight_kg']
df.loc['L. Messi',attribute]
```

```python
# get the weight and height of Messi and Ronaldo 
df.loc[players,attribute]
```

## Selecting a range with Slicing 

```python
df.head(8)
```

```python
# slice column labels
players = ['L. Messi','Cristiano Ronaldo']
df.loc[players,'age':'overall']
```

```python
columns = ['age', 'dob', 'height_cm', 'weight_kg']
df.loc['L. Messi':'V. van Dijk',columns]
```

## Selecting with Conditions

```python
# one condition: select player with height above 180cm
columns = ['age', 'dob', 'height_cm', 'weight_kg'] 
chck = df['height_cm'] > 180
df.loc[chck,columns]
```

```python
# multiple conditions : select player with height above 180 cm from Argentina
chk1 = df['height_cm']>180
chk2 = df['nationality'] == 'Argentina'
df.loc[(chk1&chk2),:]
```
