```python
import pandas as pd
```

```python
# Read Csv
df = pd.read_csv('players_20.csv')
# set new index 
df.set_index('short_name', inplace = True)
# set column
df = df[['age','dob','height_cm','weight_kg','nationality']]
```

```python
df
```

# 1 Drop Row

```python
# drop one row 
# axis parameter
# df.drop('L. Messi', axis = 0)
# index parameter
df.drop(index=['L. Messi'])
```

```python
# drop two or more rows
df.drop(index=['Cristiano Ronaldo','Neymar Jr','J. Oblak'],inplace = True)
```

```python
df
```

# 2 drop columns 

```python
# with axis 
# df.drop('dob',axis = 1)
# column  paramter
df.drop(columns=['dob'])
```

```python
# drop column by position/index 
df.columns[[-1]]
df.drop(df.columns[[-1]],axis = 1)
```

```python
# two or more columns
df.drop(columns=['dob'	,'height_cm'	,'weight_kg'],axis = 1)
```
