```python
import pandas as pd
```

```python
# reads csv file
df = pd.read_csv('players_20.csv')
# set index
df.set_index('short_name', inplace=True)
# select column
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
# show dataframe
df
```

# copy()

## deep = True

```python
# deep=True by default (Modifications to the data or indices of the copy will not be reflected in the original object )
# independent copy orignal mei changes copy mei show nhi hungi
df_copy = df.copy()
```

```python
df.loc['L. Messi','height_cm'] = 180
```

```python
df
df_copy
```

## deep = False

```python
# deep=False (Any changes to the data of the original will be reflected in the shallow copy and vice versa)
# all the changes in original dataframe will reeflected in copy dataframe
df_shallow_copy = df.copy(deep = False)
```

```python
# update value in original dataframe
df.loc['Cristiano Ronaldo','height_cm'] = 200
```

```python
# shallow copy vs original dataframe
df
df_shallow_copy
```

# Simple Assignment 

```python
# make a copy 
df_copy1 = df.copy(deep = False)
```

```python
# update value in orignal dataframe
players = ['L. Messi','Cristiano Ronaldo','Neymar Jr']
df.loc[players,'club'] = "Argentina"
```

```python
# copy vs orignal dataframe
df
df_copy1
```
