## Concatenate 
- concat() --- Concatenation ka matlab hota hai dataframes ya series ko merge ya join karna ek saath.
- must have same columns names

#### vertically concatenate

```python
df1 = pd.DataFrame({'id':['A','B','C','D'],
                   'age':[30,23,25,22]})
df2 = pd.DataFrame({'id':['E','F','G','H'],
                   'age':[40,21,19,24]})
```

```python
pd.concat([df1,df2],axis = 0, ignore_index = True)
```
