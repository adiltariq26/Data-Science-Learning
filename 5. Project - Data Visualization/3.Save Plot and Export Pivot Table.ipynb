```python
import pandas as pd 
```

```python
# reading csv file
df_population_raw = pd.read_csv('population_total.csv')
```

```python
# first look at dataset
df_population_raw
```

# Pivot_Table()
- Reshape data for easy understanding data

```python
# drop null values
df_population_raw.dropna()
```

```python
# making a pivot Table
df_pivot = df_population_raw.pivot_table(index="year",columns = "country",values = "population")
df_pivot
```

```python
# selecting some countries
df_pivot = df_pivot[['United States','India','China','Indonesia','Brazil']]
```

```python
df_pivot
```

```python
# importing matplotlib 
import matplotlib.pyplot as plt
```

# Lineplot

```python
df_pivot.plot(kind='line', xlabel = 'Year',ylabel = 'Population',
              title= 'Population From 1955-2020',figsize =(8,4))
# save plot 
plt.savefig('line_test.png')
# show plot 
plt.show()
```

```python
# export
df_pivot.to_excel('pivot_table.xlsx')
```

# Barplot

### 1-Single Barplot

```python
# selecting only one year
df_pivot_2020 = df_pivot[df_pivot.index.isin([2020])]
```

```python
# showing new dataframe
df_pivot_2020
```

```python
# transpose dataframe from rows to columns or viceversa
df_pivot_2020 = df_pivot_2020.T
```

```python
# making barplot
df_pivot_2020.plot(kind = 'bar',color = 'purple',xlabel = 'Country',ylabel = 'Population',
                   title = 'Population Bar Chart')
```

### 2-Barplot grouped by "n" variables

```python
# selecting few years
df_pivot_sample = df_pivot[df_pivot.index.isin([1980,1990,2000,2003,2005,2010,2020])]
```

```python
df_pivot_sample
```

```python
# making grouped barplot
df_pivot_sample.plot(kind = 'bar',xlabel = 'Year',ylabel = 'Population')
```

### Piechart

```python
# changing column name
df_pivot_2020.rename(columns={2020:'2020'},inplace = True)
```

```python
# making pie chart 
df_pivot_2020.plot(kind = 'pie',y='2020')
```
